```python
import shutil
import time
import os
import librosa
from librosa import display
import matplotlib.pyplot as plt
import glob
import joblib
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/My Drive/task_frames/amr.csv') #chang filename  
data
```

```python
def copy_file(src_path,video_name,emotion):
  for file in glob.glob(src_path +'*.jpg'):
    print('source path: ',file)
    name=file.split('/')
    print('/content/drive/MyDrive/frames_ravdess_label/'+emotion+'/'+video_name+'/'+name[12])
    shutil.copyfile(file,('/content/drive/MyDrive/frames_ravdess_label/'+emotion+'/'+video_name+'/'+name[12]))
```

```python
dict={0:'neutral',1:'calm',2:'happy',3:'sad',4:'angry',5:'fearful'}
#['neutral','calm', 'happy','sad','angry', 'fearful' ]
for index, row in data.iterrows():
  print(index)
  emo=dict[row['emotion']]
  old_path=row['path']
  new_path=row['file_name']
  copy_file(old_path,new_path,emo)
```
